# GPT-4o 소개
GPT-4o("omni"의 "o")는 텍스트, 오디오 및 비디오 입력의 조합을 처리하도록 설계되었으며 텍스트, 오디오 및 이미지 형식으로 출력을 생성할 수 있습니다.

### 배경
GPT-4o 이전에는 사용자가 세 가지 별도 모델로 작동하는 음성 모드를 사용하여 ChatGPT와 상호 작용할 수 있었습니다. GPT-4o는 이러한 기능을 텍스트, 비전 및 오디오 전반에 걸쳐 훈련된 단일 모델에 통합합니다. 이러한 통합 접근 방식은 텍스트, 시각, 청각 등 모든 입력이 동일한 신경망에 의해 일관되게 처리되도록 보장합니다.

### 현재 API 기능
현재 API는 `{text, image}` 입력만 지원하며 `{text}` 출력은 `gpt-4-turbo`와 동일한 형식으로 지원됩니다. 오디오를 포함한 추가 양식이 곧 도입될 예정입니다. 이 가이드는 텍스트, 이미지 및 비디오 이해를 위해 GPT-4o를 사용하는 데 도움이 됩니다.

### Process the video into two components: frames and audio

```python
import cv2
from moviepy.editor import VideoFileClip
import time
import base64

# 우리는 OpenAI DevDay Keynote Recap 비디오를 사용할 것입니다.
# 여기에서 비디오를 검토할 수 있습니다: https://www.youtube.com/watch?v=h02ti0Bl6zk
VIDEO_PATH = "data/keynote_recap.mp4"
```

```python
def process_video(video_path, seconds_per_frame=2):
    base64Frames = []
    base_video_path, _ = os.path.splitext(video_path)

    video = cv2.VideoCapture(video_path)
    total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = video.get(cv2.CAP_PROP_FPS)
    frames_to_skip = int(fps * seconds_per_frame)
    curr_frame=0

    # 비디오를 반복하고 지정된 샘플링 속도로 프레임을 추출합니다.
    while curr_frame < total_frames - 1:
        video.set(cv2.CAP_PROP_POS_FRAMES, curr_frame)
        success, frame = video.read()
        if not success:
            break
        _, buffer = cv2.imencode(".jpg", frame)
        base64Frames.append(base64.b64encode(buffer).decode("utf-8"))
        curr_frame += frames_to_skip
    video.release()

    # 비디오에서 오디오 추출
    audio_path = f"{base_video_path}.mp3"
    clip = VideoFileClip(video_path)
    clip.audio.write_audiofile(audio_path, bitrate="32k")
    clip.audio.close()
    clip.close()

    print(f"Extracted {len(base64Frames)} frames")
    print(f"Extracted audio to {audio_path}")
    return base64Frames, audio_path

# 초당 1프레임을 추출합니다. 'seconds_per_frame' 매개변수를 조정하여 샘플링 속도를 변경할 수 있습니다.
base64Frames, audio_path = process_video(VIDEO_PATH, seconds_per_frame=1)
```

```python
from IPython.display import Image, display, Audio, Markdown

## context에 맞는 프레임과 오디오 표시
display_handle = display(None, display_id=True)
for img in base64Frames:
    display_handle.update(Image(data=base64.b64decode(img.encode("utf-8")), width=600))
    time.sleep(0.025)

Audio(audio_path)
```

### 예시 1: 요약
이제 비디오 프레임과 오디오가 모두 있으므로 몇 가지 다른 테스트를 실행하여 비디오 요약을 생성하여 다양한 양식의 모델을 사용한 결과를 비교해 보겠습니다. 모델이 비디오의 전체 컨텍스트를 사용할 수 있으므로 시각적 및 오디오 입력의 컨텍스트로 생성된 요약이 가장 정확할 것으로 예상해야 합니다.

1. 시각적 요약
2. 오디오 요약
3. 시각 + 청각 요약

#### 시각적 요약
시각적 요약은 비디오의 프레임만 모델에 전송하여 생성됩니다. 프레임만 사용하면 모델이 시각적인 측면을 포착할 가능성이 높지만 발표자가 논의한 세부 사항은 놓칠 수 있습니다.

```python
response = client.chat.completions.create(
    model=MODEL,
    messages=[
    {"role": "system", "content": "비디오 요약을 생성 중입니다. 영상 요약 부탁드립니다. 마크다운으로 응답하세요."},
    {"role": "user", "content": [
        "영상의 프레임들입니다.",
        *map(lambda x: {"type": "image_url",
                        "image_url": {"url": f'data:image/jpg;base64,{x}', "detail": "low"}}, base64Frames)
        ],
    }
    ],
    temperature=0,
)
print(response.choices[0].message.content)
```

결과는 예상한 대로입니다. 모델은 비디오 영상의 높은 수준의 측면을 캡처할 수 있지만 연설에서 제공되는 세부 정보는 놓칩니다.

#### 오디오 요약
오디오 요약은 모델에 오디오 대본을 전송하여 생성됩니다. 오디오만 사용하면 모델이 오디오 콘텐츠 쪽으로 편향될 가능성이 높으며 프리젠테이션과 시각적 자료가 제공하는 컨텍스트를 놓칠 수 있습니다.

GPT-4o용 `{audio}` 입력은 현재 사용할 수 없지만 곧 제공될 예정입니다! 지금은 기존 'whisper-1' 모델을 사용하여 오디오를 처리합니다.

```python
# 오디오를 텍스트로 변환
transcription = client.audio.transcriptions.create(
    model="whisper-1",
    file=open(audio_path, "rb"),
)
## 선택사항: 스크립트를 인쇄하려면 아래 줄의 주석 처리를 해제하세요.
#print("Transcript: ", transcription.text + "\n\n")

response = client.chat.completions.create(
    model=MODEL,
    messages=[
    {"role": "system", "content":"""텍스트 변환 요약을 생성 중입니다. 제공된 텍스트 변환의 요약을 만듭니다. 마크다운 형식의 한국어로 응답."""},
    {"role": "user", "content": [
        {"type": "text", "text": f"오디오의 텍스트 변환은: {transcription.text}"}
        ],
    }
    ],
    temperature=0,
)
print(response.choices[0].message.content)
```

오디오 요약은 연설 중에 논의된 내용에 치우치지만, 비디오 요약보다는 구조가 훨씬 덜 명확하게 나옵니다.

#### 오디오 + 비주얼 요약
오디오 + 비주얼 요약은 비디오의 시각적 정보와 오디오를 동시에 모델에 보내어 생성됩니다. 이 두 가지를 모두 보내면 모델은 비디오 전체를 한 번에 인식할 수 있기 때문에 더 나은 요약을 제공할 것으로 예상됩니다.

```python
## Generate a summary with visual and audio
response = client.chat.completions.create(
    model=MODEL,
    messages=[
    {"role": "system", "content":"""비디오 요약을 생성 중입니다. 제공된 비디오의 요약과 해당 내용을 작성합니다.  한국어 마크다운으로 응답"""},
    {"role": "user", "content": [
        "영상의 프레임들입니다.",
        *map(lambda x: {"type": "image_url",
                        "image_url": {"url": f'data:image/jpg;base64,{x}', "detail": "low"}}, base64Frames),
        {"type": "text", "text": f"The audio transcription is: {transcription.text}"}
        ],
    }
],
    temperature=0,
)
print(response.choices[0].message.content)
```

비디오와 오디오를 결합한 후, 우리는 비디오의 시각적 요소와 오디오 요소 모두에서 정보를 사용하여 이벤트에 대한 훨씬 더 상세하고 포괄적인 요약을 얻을 수 있습니다.

### 예시 2: 질문과 답변
Q&A의 경우, 앞서 사용한 동일한 개념을 사용하여 처리된 비디오에 질문을 하고, 입력 모달리티를 결합하는 것의 이점을 보여주기 위해 동일한 3가지 테스트를 실행합니다:
1. 시각적 Q&A
2. 오디오 Q&A
3. 시각 + 오디오 Q&A

```python
QUESTION = "질문: 샘 알트만은 왜 창문을 올리고 라디오를 켜는 것에 대한 예를 들었나요?"
```

```python
qa_visual_response = client.chat.completions.create(
    model=MODEL,
    messages=[
    {"role": "system", "content": "비디오를 사용하여 제공된 질문에 답하십시오. 한국어 마크다운으로 응답."},
    {"role": "user", "content": [
        "영상의 프레임들입니다.",
        *map(lambda x: {"type": "image_url", "image_url": {"url": f'data:image/jpg;base64,{x}', "detail": "low"}}, base64Frames),
        QUESTION
        ],
    }
    ],
    temperature=0,
)
print("Visual QA:\n" + qa_visual_response.choices[0].message.content)
```

```python
qa_audio_response = client.chat.completions.create(
    model=MODEL,
    messages=[
    {"role": "system", "content":"""제공된 질문에 답하려면 audio transcription을 사용하세요. 한국어 마크다운으로 응답하세요."""},
    {"role": "user", "content": f"The audio transcription is: {transcription.text}. \n\n {QUESTION}"},
    ],
    temperature=0,
)
print("Audio QA:\n" + qa_audio_response.choices[0].message.content)
```

```python
qa_both_response = client.chat.completions.create(
    model=MODEL,
    messages=[
    {"role": "system", "content":"""비디오와 스크립트를 사용하여 제공된 질문에 답하세요. 한국어로 답하세요."""},
    {"role": "user", "content": [
        "These are the frames from the video.",
        *map(lambda x: {"type": "image_url",
                        "image_url": {"url": f'data:image/jpg;base64,{x}', "detail": "low"}}, base64Frames),
                        {"type": "text", "text": f"The audio transcription is: {transcription.text}"},
        QUESTION
        ],
    }
    ],
    temperature=0,
)
print("Both QA:\n" + qa_both_response.choices[0].message.content)
```

세 가지 답변을 비교하면 비디오의 오디오와 시각적 자료를 모두 사용하여 가장 정확한 답변이 생성됩니다. Sam Altman은 기조 연설 중에 창문이나 라디오를 올리는 것에 대해 논의하지 않았지만 00:39 위치에 Sam의 배경 화면으로 그러한 내용의 예제가 표시되는 동안 모델이 단일 요청으로 여러 기능을 실행할 수 있는 향상된 기능을 언급했습니다.
